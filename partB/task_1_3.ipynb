## Main Baseline

The primary baseline the paper compares against is **Cost-Sensitive One-Versus-All SVM (CS-OVA)**. This is a standard approach where K binary SVM classifiers are trained (one per class), with each SVM using class-dependent misclassification weights derived from the cost matrix. The CS-OVA method was a well-established technique for cost-sensitive multiclass classification before OSSVR and is explicitly used as the main comparison throughout the paper's experiments.

## Limitation of the Baseline (CS-OVA)

The paper identifies that CS-OVA uses a **weighted hinge loss** to incorporate costs — it multiplies the hinge loss by a cost-derived weight for each training example. However, the hinge loss fundamentally encodes only a binary signal: whether the prediction is correct or incorrect. It cannot distinguish between a *slightly wrong* prediction (low cost) and a *catastrophically wrong* prediction (high cost). The cost information is reduced to a scalar weight rather than being embedded in the structure of the learning target. This means CS-OVA treats all incorrect predictions for a given class as equally bad (up to a scalar multiplier), losing the ordinal cost structure that distinguishes different types of errors.

## How OSSVR Overcomes This Limitation

OSSVR addresses this by replacing the classification loss entirely with a **regression loss on the actual cost values** (specifically, the regret vector). Instead of learning "correct vs. incorrect" with a weight, each regressor directly approximates the cost of predicting a particular class. This preserves the magnitude and ordering of costs in the learning target, allowing the model to allocate its capacity toward distinguishing between cheap and expensive errors.

## Scenario Where OSSVR Would NOT Outperform CS-OVA

OSSVR's advantage lies in exploiting non-uniform cost structure — it benefits most when different types of misclassification errors have very different costs. **When the cost matrix is uniform** (all off-diagonal entries are equal, i.e., C[y, k] = c for all y ≠ k), the regret vector for any example becomes a binary vector: 0 for the correct class and c for every incorrect class. In this case, the regression targets carry no more information than a standard one-hot label scaled by a constant. The OSSVR regressors are effectively solving a harder problem (regression) to achieve the same result that a simpler binary classifier (CS-OVA) can achieve directly. Moreover, the regression formulation introduces additional hyperparameters (e.g., the ε parameter in SVR) that can make tuning harder without any compensating benefit. Therefore, under uniform costs, CS-OVA is expected to be at least as good as OSSVR while being simpler and faster to train.

This is not explicitly discussed by the authors, but it follows logically from the method's design: the advantage of preserving cost magnitudes disappears when all costs are identical.
